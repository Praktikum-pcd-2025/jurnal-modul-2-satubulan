```python
def gabung_vertikal(img1, img2, img3, img4, img5, img6):
    tinggi1 = len(img1)
    tinggi2 = len(img2)
    tinggi3 = len(img3)
    tinggi4 = len(img4)
    tinggi5 = len(img5)
    tinggi6 = len(img6)
    lebar1 = len(img1[0])
    lebar2 = len(img2[0])
    lebar3 = len(img3[0])
    lebar4 = len(img4[0])
    lebar5 = len(img5[0])
    lebar6 = len(img6[0])
    lebar_max = max(lebar1, lebar2, lebar3, lebar4, lebar5, lebar6)

    result = [[0 for _ in range(lebar_max)] for _ in range(tinggi1 + tinggi2+ tinggi3 + tinggi4+ tinggi5+tinggi6)]

    for i in range(tinggi1):
        for j in range(lebar1):
            result[i][j] = img1[i][j]
    for i in range(tinggi2):
        for j in range(lebar2):
            result[tinggi1 + i][j] = img2[i][j]
    for i in range(tinggi3):
        for j in range(lebar3):
            result[tinggi2 + i][j] = img3[i][j]
    for i in range(tinggi4):
        for j in range(lebar4):
            result[tinggi3 + i][j] = img4[i][j]
    for i in range(tinggi5):
        for j in range(lebar5):
            result[tinggi4 + i][j] = img5[i][j]
    for i in range(tinggi6):
        for j in range(lebar6):
            result[tinggi5 + i][j] = img6[i][j]

    return result
```

```python
tokohmasyarakat= gabung_vertikal(badeni, albert, rafal, jalenta, oczy, draka)
plt.imshow(tokohmasyarakat)
print(tokohmasyarakat.shape)
```
